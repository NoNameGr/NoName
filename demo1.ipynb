```python
import tkinter as tk    
from tkinter import font as tkfont
from tkinter import *
from PIL import ImageTk, Image
import os
from tkinter import filedialog
```

```python
class SampleApp(tk.Tk): 

    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)
        self.geometry('500x500')
        self.title('Duoi hinh bat chu')
 
        self.title_font = tkfont.Font(family='Helvetica', size=18, weight="bold", slant="italic")
 
        # the container is where we'll stack a bunch of frames
        # on top of each other, then the one we want visible
        # will be raised above the others
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        container.grid_rowconfigure(1, weight=1)
        container.grid_columnconfigure(1, weight=1)
 
        self.frames = {}
        for F in (StartPage, TiengViet, English) : 
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
 
            # put all of the pages in the same location;
            # the one on the top of the stacking order
            # will be the one that is visible.
            frame.grid(row=0, column=0, sticky="nsew")
 
        self.show_frame("StartPage")
    
 
    def show_frame(self, page_name):
        '''Show a frame for the given page name'''
        frame = self.frames[page_name]
        frame.tkraise()
 
    
class StartPage(tk.Frame) : 
    
    def __init__(self, parent, controller) : 
        tk.Frame.__init__(self, parent)
        self.controller = controller
        label = tk.Label(self, text='Duoi hinh bat chu', font=controller.title_font)
        label.pack(side='top,middle'
        
        
        
        button1 = tk.Button(self, text='Tieng viet',
            comman=lambda: controller.show_frame('TiengViet'))
        button2 = tk.Button(self, text='English',
            comman=lambda: controller.show_frame('English'))
        button1.pack()
        button2.pack()
        
class TiengViet(tk.Frame) : 
    
    def __init__(self, parent, controller) : 
        tk.Frame.__init__(self, parent)
        self.controller = controller
        label = tk.Label(self, text='Hay chon muc do choi', font=controller.title_font)
        label.pack(side='top', fill='x', pady=10)
        
        button1 = tk.Button(self, text='De',
            command=lambda: controller.show_frame('De'))
        button2 = tk.Button(self, text='Trung binh',
            command=lambda: controller.show_frame('TrungBinh'))
        button3 = tk.Button(self, text='Kho',
            command=lambda: controller.show_frame('Kho'))
        button1.pack()
        button2.pack()
        button3.pack()
        
        button = tk.Button(self, text="Quay tro lai trang dau tien", 
          command=lambda: controller.show_frame("StartPage")) 
        button.pack()
class English(tk.Frame) : 
    
     def __init__(self, parent, controller) : 
        tk.Frame.__init__(self, parent)
        self.controller = controller
        label = tk.Label(self, text='Choose the level', font=controller.title_font)
        label.pack(side='top', fill='x', pady=10)
        
        button1 = tk.Button(self, text='Easy',
            command=lambda: controller.show_frame('Easy'))
        button2 = tk.Button(self, text='Medium',
            command=lambda: controller.show_frame('Medium'))
        button3 = tk.Button(self, text='Hard',
            command=lambda: controller.show_frame('Hard'))
        button1.pack()
        button2.pack()
        button3.pack()
        
        
        button = tk.Button(self, text="Go to the start page", 
                  command=lambda: controller.show_frame("StartPage")) 
        button.pack()
```

```python
if __name__ == "__main__": 
    app = SampleApp() 
    app.mainloop() 
    app_root.mainloop()
     
```
